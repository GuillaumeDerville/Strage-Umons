```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from scipy.stats import kurtosis, skew
from sklearn.preprocessing import RobustScaler
import tensorflow as tf
from tensorflow import keras
import time

# ════════════════════════════════════════════════════════════════════════════
# PARAMÈTRES À ADAPTER
# ════════════════════════════════════════════════════════════════════════════

SAMPLING_RATE = 25_600   # Hz du capteur
WINDOW_SIZE   = 1024     # samples par fenêtre
WINDOW_STEP   = 512      # pas de glissement

# Durées en secondes
WARMUP_DURATION  = 30 * 60   # 30 min de rodage
COLLECT_DURATION = 90 * 60   # 1h30 de collecte

K_SIGMA      = 3.0   # seuil = mean + K_SIGMA × std
VOTE_WINDOW  = 10    # fenêtre vote
VOTE_THRESH  = 5     # ≥5/10 pour confirmer

print("Paramètres chargés ✓")
print(f"  Rodage   : {WARMUP_DURATION//60} min")
print(f"  Collecte : {COLLECT_DURATION//60} min")
print(f"  Seuil    : {K_SIGMA}σ | Vote : ≥{VOTE_THRESH}/{VOTE_WINDOW}")
```

```python
FEATURE_NAMES = [
    "mean", "std", "rms", "max", "min", "peak_to_peak",
    "skewness", "kurtosis", "crest_factor", "shape_factor",
    "impulse_factor", "freq_dominant", "spectral_energy", "spectral_entropy",
]
LOG_FEATURES = ["freq_dominant", "spectral_energy"]

def extract_features(window, fs=SAMPLING_RATE):
    eps = 1e-10
    n = len(window)
    mean_val     = np.mean(window)
    std_val      = np.std(window)
    rms_val      = np.sqrt(np.mean(window**2))
    max_val      = np.max(np.abs(window))
    min_val      = np.min(np.abs(window))
    p2p_val      = np.ptp(window)
    skew_val     = float(skew(window))
    kurt_val     = float(kurtosis(window))
    mean_abs     = np.mean(np.abs(window))
    crest_val    = max_val / (rms_val + eps)
    shape_val    = rms_val / (mean_abs + eps)
    impulse_val  = max_val / (mean_abs + eps)
    fft_mag      = np.abs(np.fft.rfft(window))
    freqs        = np.fft.rfftfreq(n, d=1.0/fs)
    freq_dom     = freqs[np.argmax(fft_mag)]
    spec_energy  = np.sum(fft_mag**2)
    psd_norm     = fft_mag**2 / (spec_energy + eps)
    spec_entropy = -np.sum(psd_norm * np.log2(psd_norm + eps))
    return np.array([mean_val, std_val, rms_val, max_val, min_val, p2p_val,
                     skew_val, kurt_val, crest_val, shape_val, impulse_val,
                     freq_dom, spec_energy, spec_entropy], dtype=np.float32)

def apply_log(feat):
    out = feat.copy()
    for name in LOG_FEATURES:
        idx = FEATURE_NAMES.index(name)
        out[..., idx] = np.log10(np.abs(feat[..., idx]) + 1e-10)
    return out

def preprocess(feat_array, scaler=None, fit=False):
    feat_log = apply_log(feat_array)
    if fit:
        scaler = RobustScaler()
        X = scaler.fit_transform(feat_log)
    else:
        X = scaler.transform(feat_log.reshape(1, -1) if feat_log.ndim == 1 else feat_log)
    return np.clip(X, -5, 5), scaler

print("Fonctions core prêtes ✓")
```

```python
def phase_warmup(data_stream):
    """
    Phase 1 — Rodage pendant WARMUP_DURATION secondes.
    Consomme le flux sans rien calculer.

    data_stream : générateur qui yield des chunks numpy 1D en continu
    """
    print("═"*60)
    print(f"  PHASE 1 — RODAGE ({WARMUP_DURATION//60} min)")
    print("═"*60)

    t0 = time.time()
    while time.time() - t0 < WARMUP_DURATION:
        remaining = WARMUP_DURATION - (time.time() - t0)
        next(data_stream)   # consomme sans traiter
        print(f"\r  ⏳ {int(remaining//60):02d}:{int(remaining%60):02d} restantes   ",
              end="", flush=True)

    print(f"\n  ✓ Rodage terminé\n")


def phase_collect(data_stream):
    """
    Phase 2 — Collecte pendant COLLECT_DURATION secondes.
    Extrait les features de chaque chunk reçu.

    Retourne : features_raw (np.array shape [N, 14])
    """
    print("═"*60)
    print(f"  PHASE 2 — COLLECTE ({COLLECT_DURATION//60} min)")
    print("═"*60)

    features_list = []
    buffer = []
    t0 = time.time()

    while time.time() - t0 < COLLECT_DURATION:
        remaining = COLLECT_DURATION - (time.time() - t0)
        chunk = next(data_stream)
        buffer.extend(chunk.tolist())

        # Extrait les fenêtres disponibles
        while len(buffer) >= WINDOW_SIZE:
            window = np.array(buffer[:WINDOW_SIZE], dtype=np.float32)
            buffer = buffer[WINDOW_STEP:]
            feat = extract_features(window)
            features_list.append(feat)

        print(f"\r  📡 {int(remaining//60):02d}:{int(remaining%60):02d} restantes"
              f" | {len(features_list)} fenêtres extraites   ",
              end="", flush=True)

    features_raw = np.array(features_list, dtype=np.float32)
    print(f"\n  ✓ {features_raw.shape[0]} vecteurs de features extraits\n")
    return features_raw


def phase_train(features_raw):
    """
    Entraîne l'autoencodeur sur les features normales.
    Retourne : model, scaler, threshold, errors, history
    """
    print("═"*60)
    print("  ENTRAÎNEMENT — Autoencodeur")
    print("═"*60)

    X_train, scaler = preprocess(features_raw, fit=True)

    inp   = keras.Input(shape=(14,))
    enc   = keras.layers.Dense(8,  activation="relu")(inp)
    enc   = keras.layers.Dense(4,  activation="relu")(enc)
    dec   = keras.layers.Dense(8,  activation="relu")(enc)
    out   = keras.layers.Dense(14, activation="linear")(dec)
    model = keras.Model(inp, out)
    model.compile(optimizer="adam", loss="mse")

    history = model.fit(
        X_train, X_train,
        epochs=100, batch_size=32,
        validation_split=0.1,
        callbacks=[keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)],
        verbose=0
    )

    recon     = model.predict(X_train, verbose=0)
    errors    = np.mean((X_train - recon)**2, axis=1)
    mean_err  = float(np.mean(errors))
    std_err   = float(np.std(errors))
    threshold = mean_err + K_SIGMA * std_err

    print(f"  Epochs : {len(history.history['loss'])}"
          f" | Loss : {history.history['loss'][-1]:.5f}"
          f" | Val  : {history.history['val_loss'][-1]:.5f}")
    print(f"  Seuil  : {threshold:.5f}  "
          f"(mean={mean_err:.5f} + {K_SIGMA}σ={K_SIGMA*std_err:.5f})")
    print(f"  ✓ Modèle prêt\n")

    return model, scaler, threshold, errors, history


def phase_detect(data_stream, model, scaler, threshold):
    """
    Phase 3 — Détection en continu avec vote majoritaire.
    S'arrête quand data_stream est épuisé ou manuellement (Ctrl+C).

    Retourne : detect_errors, detect_labels, detect_smooth,
               first_anomaly_raw, first_anomaly_confirmed
    """
    print("═"*60)
    print("  PHASE 3 — DÉTECTION TEMPS RÉEL")
    print(f"  Vote : ≥{VOTE_THRESH}/{VOTE_WINDOW} pour confirmer")
    print("═"*60)

    detect_errors           = []
    detect_labels           = []
    detect_smooth           = []
    first_anomaly_raw       = None
    first_anomaly_confirmed = None
    recent                  = []
    buffer                  = []
    window_idx              = 0

    try:
        for chunk in data_stream:
            buffer.extend(chunk.tolist())

            while len(buffer) >= WINDOW_SIZE:
                window = np.array(buffer[:WINDOW_SIZE], dtype=np.float32)
                buffer = buffer[WINDOW_STEP:]
                window_idx += 1

                feat = extract_features(window)
                X, _ = preprocess(feat, scaler=scaler, fit=False)
                recon = model.predict(X, verbose=0)
                err   = float(np.mean((X - recon)**2))
                anomaly = err > threshold

                detect_errors.append(err)
                detect_labels.append(int(anomaly))

                # Vote glissant
                recent.append(int(anomaly))
                if len(recent) > VOTE_WINDOW:
                    recent.pop(0)
                confirmed = sum(recent) >= VOTE_THRESH
                detect_smooth.append(int(confirmed))

                if anomaly and first_anomaly_raw is None:
                    first_anomaly_raw = window_idx

                if confirmed and first_anomaly_confirmed is None:
                    first_anomaly_confirmed = window_idx
                    print(f"\n  🔴 ANOMALIE CONFIRMÉE fenêtre #{window_idx}"
                          f" | erreur={err:.4f}")

                status = ("🔴 CONFIRMÉE " if confirmed else
                          "⚠️  détectée  " if anomaly else
                          "🟢 normal    ")
                print(f"\r  {status} | fenêtre #{window_idx:<6} | "
                      f"err={err:.4f} | vote={sum(recent)}/{min(len(recent),VOTE_WINDOW)}   ",
                      end="", flush=True)

    except (StopIteration, KeyboardInterrupt):
        print(f"\n  ✓ Détection arrêtée")

    n_raw       = sum(detect_labels)
    n_confirmed = sum(detect_smooth)
    total       = len(detect_labels)

    print(f"\n{'═'*60}")
    print(f"  RÉSULTATS")
    print(f"{'═'*60}")
    print(f"  Fenêtres analysées      : {total}")
    print(f"  Anomalies brutes        : {n_raw}  ({100*n_raw/max(total,1):.1f}%)")
    print(f"  Anomalies confirmées    : {n_confirmed}  ({100*n_confirmed/max(total,1):.1f}%)")
    print(f"  1ère anomalie brute     : fenêtre #{first_anomaly_raw}")
    print(f"  1ère anomalie confirmée : fenêtre #{first_anomaly_confirmed}")

    return (detect_errors, detect_labels, detect_smooth,
            first_anomaly_raw, first_anomaly_confirmed)

print("Fonctions pipeline prêtes ✓")
```

```python
def plot_results(history, errors, threshold,
                 detect_errors, detect_smooth,
                 first_anomaly_raw, first_anomaly_confirmed,
                 title="Bearing"):

    mean_err    = float(np.mean(errors))
    n_confirmed = sum(detect_smooth)
    total       = len(detect_smooth)
    idx         = list(range(1, total + 1))

    fig, axes = plt.subplots(3, 1, figsize=(14, 10))

    axes[0].plot(history.history['loss'],     label="Train", color="steelblue")
    axes[0].plot(history.history['val_loss'], label="Val",   color="orange")
    axes[0].set_title("Entraînement — données normales", fontsize=11)
    axes[0].set_xlabel("Epoch")
    axes[0].set_ylabel("MSE")
    axes[0].legend()
    axes[0].spines[['top','right']].set_visible(False)

    axes[1].plot(errors, color="steelblue", alpha=0.7, linewidth=0.8)
    axes[1].axhline(threshold, color="red",    linestyle="--", linewidth=1.5,
                    label=f"Seuil = {threshold:.4f}")
    axes[1].axhline(mean_err,  color="orange", linestyle="--", linewidth=1,
                    label=f"Moyenne = {mean_err:.4f}")
    axes[1].set_title("Erreurs reconstruction — phase normale", fontsize=11)
    axes[1].set_ylabel("MSE")
    axes[1].set_xlabel("Fenêtre")
    axes[1].legend()
    axes[1].spines[['top','right']].set_visible(False)

    colors_d = ["tomato" if a else "steelblue" for a in detect_smooth]
    axes[2].scatter(idx, detect_errors, c=colors_d, s=10, alpha=0.7)
    axes[2].axhline(threshold, color="red", linestyle="--", linewidth=1.5,
                    label=f"Seuil = {threshold:.4f}")
    if first_anomaly_confirmed:
        axes[2].axvline(first_anomaly_confirmed, color="black",
                        linestyle=":", linewidth=1.5,
                        label=f"1ère confirmée : #{first_anomaly_confirmed}")
    if first_anomaly_raw:
        axes[2].axvline(first_anomaly_raw, color="gray",
                        linestyle=":", linewidth=1,
                        label=f"1ère brute : #{first_anomaly_raw}")
    axes[2].set_yscale('log')
    axes[2].set_title(
        f"Détection — {title} | {n_confirmed}/{total} confirmées "
        f"(vote ≥{VOTE_THRESH}/{VOTE_WINDOW})", fontsize=11)
    axes[2].set_xlabel("Fenêtre")
    axes[2].set_ylabel("Erreur MSE (log)")

    from matplotlib.patches import Patch
    axes[2].legend(handles=[
        Patch(color="steelblue", label="Normal"),
        Patch(color="tomato",    label="Anomalie confirmée"),
        plt.Line2D([0],[0], color="red",   linestyle="--",
                   label=f"Seuil = {threshold:.4f}"),
        plt.Line2D([0],[0], color="black", linestyle=":",
                   label=f"1ère confirmée : #{first_anomaly_confirmed}"),
        plt.Line2D([0],[0], color="gray",  linestyle=":",
                   label=f"1ère brute : #{first_anomaly_raw}"),
    ])
    axes[2].spines[['top','right']].set_visible(False)

    plt.tight_layout()
    plt.savefig(f"detection_{title}.png", dpi=150, bbox_inches="tight")
    plt.show()
    print(f"Sauvegardé → detection_{title}.png ✓")

print("Fonction plot_results prête ✓")
```
