```python
import sys
sys.path.insert(0, '/home/xilinx/jupyter_notebooks')

from axi_stream_driver import NeuralNetworkOverlay
import numpy as np

# Charger les paramètres du scaler
center = np.load("/home/xilinx/jupyter_notebooks/getting_started/scaler_center.npy")
scale  = np.load("/home/xilinx/jupyter_notebooks/getting_started/scaler_scale.npy")

# Charger les données sans la colonne d'index
df_raw = np.genfromtxt("/home/xilinx/jupyter_notebooks/getting_started/features_set2_bearing1.csv", 
                        delimiter=',', skip_header=1)
df_raw = df_raw[:, 1:]  # Enlever l'index
X = ((df_raw - center) / scale).astype(np.float32)
print(f"Données chargées : {X.shape}")

# Charger l'overlay
nn = NeuralNetworkOverlay(
    "/home/xilinx/jupyter_notebooks/getting_started/myproject.bit",
    X[0:1].shape, X[0:1].shape
)
print("Overlay chargé !")

# Test sur un seul échantillon
y_hw, latency, throughput = nn.predict(X[0:1], profile=True)
print(f"Latence : {latency:.2f} µs")
print(f"Sortie  : {y_hw}")
```

```python
# Inférence sur tout le dataset
y_hw, latency, throughput = nn.predict(X, profile=True)

# Calcul erreur de reconstruction
errors = np.mean((X - y_hw)**2, axis=1)

# Seuil
threshold = 1.8729

# Détection
detected = np.where(errors > threshold)[0]
print(f"Premier fichier détecté : #{detected[0]+1 if len(detected) > 0 else 'aucun'}")
print(f"Fausses alarmes (fichiers 1-450) : {np.sum(errors[:450] > threshold)}")
print(f"Latence moyenne : {latency:.2f} µs")
print(f"Débit : {throughput:.0f} inférences/s")
```

```python
WARMUP = 15
TRAIN_END = 450

# Exclure le warmup
max_sain = errors[WARMUP:TRAIN_END].max()
new_threshold = max_sain * 1.5
print(f"Max erreur phase saine (sans warmup) : {max_sain:.4f}")
print(f"Nouveau seuil : {new_threshold:.4f}")

detected = np.where(errors[WARMUP:] > new_threshold)[0]
print(f"Premier fichier détecté : #{detected[0] + WARMUP + 1 if len(detected) > 0 else 'aucun'}")
print(f"Fausses alarmes (fichiers {WARMUP+1}-{TRAIN_END}) : {np.sum(errors[WARMUP:TRAIN_END] > new_threshold)}")
```

```python
from pynq import allocate
import matplotlib.pyplot as plt
import time
import numpy as np

plt.figure(figsize=(12,4))
plt.semilogy(errors, label='Erreur FPGA')
plt.axhline(y=new_threshold, color='r', linestyle='--', label=f'Seuil FPGA {new_threshold:.4f}')
plt.axvline(x=WARMUP, color='g', linestyle=':', label='Fin warmup')
plt.axvline(x=TRAIN_END, color='orange', linestyle=':', label='Fin entraînement')
plt.axvline(x=detected[0]+WARMUP, color='purple', linestyle='-', label=f'Détection #{detected[0]+WARMUP+1}')
plt.xlabel('Fichier #')
plt.ylabel('Erreur reconstruction (log)')
plt.title("Détection d'anomalie FPGA — Bearing 1")
plt.legend()
plt.savefig('/home/xilinx/jupyter_notebooks/getting_started/fpga_detection_finale.png')
plt.show()

# Mesure de latence
N_REPEAT = 100
times = []

input_buf = allocate(shape=(14,), dtype=np.float32)
output_buf = allocate(shape=(14,), dtype=np.float32)

for _ in range(N_REPEAT):
    t0 = time.perf_counter()
    input_buf[:] = X[0]
    nn.dma.sendchannel.transfer(input_buf)
    nn.dma.recvchannel.transfer(output_buf)
    nn.dma.sendchannel.wait()
    nn.dma.recvchannel.wait()
    t1 = time.perf_counter()
    times.append((t1 - t0) * 1e6)

times = np.array(times)
print(f"Latence FPGA (DMA inclus) :")
print(f"  Moyenne  : {times.mean():.2f} µs")
print(f"  Min      : {times.min():.2f} µs")
print(f"  Max      : {times.max():.2f} µs")
print(f"  Std      : {times.std():.2f} µs")
print(f"  Médiane  : {np.median(times):.2f} µs")
print(f"Latence calcul HLS (prédite par synthèse) : 1.56 µs")
print(f"Latence totale mesurée (DMA + calcul)     : {times.mean():.2f} µs")
print(f"Overhead DMA                               : {times.min() - 1.56:.2f} µs")
```
